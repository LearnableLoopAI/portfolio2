## 2 SYNTHESIZE DATA

```python
df = pd.DataFrame(columns=[
    'pat_last', 'pat_first', 'pat_street', 'pat_city', 'pat_zip', 'pat_state', 'pat_dob', #'pat_gender', #PATIENT
    'enc_date', 'enc_practitioner_last', 'enc_practitioner_first', 'enc_practitioner_role', 'enc_type', #ENCOUNTER
    'con_name', 'con_diag_date', #CONDITION
    'med_name' #MEDICATION
])
dp = []
for i in range(10000):
    #PATIENT
    #hardcoded for now - need to sample from list
    dp = [sample(LAST_NAMES, 1)[0]]+[sample(FIRST_NAMES, 1)[0]]+[sample(STREET_NAMES, 1)[0]]+[sample(CITIES, 1)[0]]+['98230']+['WA']+['2007-01-01'] #['M']+ #PATIENT
    #ENCOUNTER
    #hardcoded for now - need to sample from list
    dp = dp + ['2021-07-22']+[sample(LAST_NAMES, 1)[0]]+[sample(FIRST_NAMES, 1)[0]]+['Physician/RN/PA/etc']+['FirstVisit/Checkup/Refil/etc'] #ENCOUNTER    
    #CONDITION
    con_med_sample = df_con_med.sample()
    dp = dp + [con_med_sample['disease'].values[0]]+['2001-12-04'] #CONDITION
    #MEDICATION    
    dp = dp + [con_med_sample['drug'].values[0]] #MEDICATION
    #add new datapoint to dataframe
    df.loc[i] = dp
```
